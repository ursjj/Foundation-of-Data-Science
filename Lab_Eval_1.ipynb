```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('https://github.com/flaviovdf/evcomp2018/raw/master/data/educ_inc.csv')
df.head()
```

```python
#1.Group the table by Educational Attainment and sum the Population Count in each category(2).

df.groupby('Educational Attainment')['Population Count'].sum()
```

```python
df.groupby('Age').count()
```

```python
#2. Analyse the percentage distribution of educational attainment among adult Californians(2).

df[(df['Age']=='65 to 80+')].groupby('Educational Attainment')['Population Count'].sum()/df.groupby('Educational Attainment')['Population Count'].sum()*100
```

```python
#3. Using pivot , get a contingency table (a table of counts) of adult Californians cross-classified by Educational Attainment and Personal Income(2).

contingency_table = pd.pivot_table(df,index='Educational Attainment', columns='Personal Income', aggfunc=len)
print(contingency_table)
```
